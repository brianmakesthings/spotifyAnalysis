### **4.1 Represention of Lyrics Features**

To better compute a score for lyrics pertaining to a specific song in a playlist as a part of an entire playlist' lyrics dataset, we introduce the Term Frequency-Inverse Document Frequency (or TFIDF for short). The TFIDF score depicts the number of occurences of a word of a lyric in relation to an entire list of words throughout a lyrics document [1].

We begin by determining the TF score for each word within a particular lyric using the formula below:

$$ 
TF_{i,j} = \frac{|number\space \,of\space \,occurences\space \,of\space \,word\space \,i\space \,within\space \,a\space \,lyric\space \,j|}{|lyric\space \,j|}\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\space\,[1]
$$

Next, we calculate the inverse document frequenccy score for each lyric within the document

$$ 
IDF_{i} = \frac{|total\space\,number\space \,of\space \,lyrics\space \,in\space \,dataset\space \,j|}{|number\space\,of\space \,lyrics\space \,containing\space \,word\space\,i|}\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace[1]$$

The resulting TFIDF score from the computations becomes:

$\enspace\,TFIDF_{i,j} = TF_{i,j}\space\,\times\,\space\,IDF_{i}\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,\enspace\,[1]$
