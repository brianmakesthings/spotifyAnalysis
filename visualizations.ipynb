```python
import numpy as np
import pandas as pd
```

```python
corpus = pd.read_csv("spotifyData.csv", index_col=0)
corpus.head()
```

```python
playlist = pd.read_csv("playlists/spotifySadBops.csv", index_col=0).drop("Unnamed: 0", axis=1)
playlist.head()
```

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
numericColumns = ["danceability",
                  "energy",
                  "loudness",
                  "speechiness",
                  "acousticness",
                  "instrumentalness",
                  "liveness",
                  "valence",
                  "tempo"]
comparisonDf = pd.concat([corpus[numericColumns], playlist[numericColumns]], keys=['corpus', 'playlist'])\
    .reset_index(1, drop=True)\
    .reset_index(col_fill='Source')
comparisonDf
fig = px.scatter_matrix(comparisonDf, 
    dimensions=numericColumns,
    color='index',
    width=1500,
    height=1500)
fig.show()
fig.write_html("./sadSongsVsCorpus.html")
```
